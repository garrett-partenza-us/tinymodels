```python
from tinygrad.tensor import Tensor
from tinygrad.nn import Conv2d
```

```python
conv = Conv2d(3, 3, 3, padding=1)
```

```python
def upsample(x):
    bs,c,py,px = x.shape
    x = x.reshape(bs, c, py, 1, px, 1).expand(bs, c, py, 2, px, 2).reshape(bs, c, py*2, px*2)
    return x, conv(x)
```

```python
def _phase_shift(inpt, scale):
    b, w, h, _ = inpt.shape
    X = inpt.transpose(1, 2, 3, 0).reshape(w, h, scale, scale, b)
    X = np.concatenate(X, axis=1)
    X = np.concatenate(X, axis=1)
    X = X.transpose(2, 0, 1)
    return np.reshape(X, (b, w * scale, h * scale, 1))
```

```python
# https://nico-curti.github.io/NumPyNet/NumPyNet/layers/pixelshuffle_layer.html

inpt = np.random.randint(255, size=(8, 96, 96, 16))
print(_phase_shift(inpt, 4).shape)

inpt = np.random.randint(255, size=(8, 96, 96, 48))

scale = 4
batch, w, h, c = inpt.shape
channel_output = c // scale**2 
output = np.concatenate(
    [_phase_shift(inpt[:, :, :, range(i, c, channel_output)], scale) for i in range(channel_output)],
    axis=3
)

output.shape
```
