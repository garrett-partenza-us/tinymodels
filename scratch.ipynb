```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = np.load("vit/losstrain.npy")
test = np.load("vit/losseval.npy")
```

```python
plt.figure(figsize=(12,6))
plt.title("MSE Loss")
plt.plot(np.convolve(train, np.ones(10)/10, mode='valid'), label='batch train loss (rolling)')
test_x, test_y = zip(*(((x)*100,y) for x,y in enumerate(test)))
plt.scatter(test_x, test_y, color='red', label='eval loss')
plt.legend()
plt.savefig("loss.png")
```
